```python
keywords = []
f = open('data/keywords.txt', 'r')
lines = f.readlines()
f.close()
for line in lines:
    a,b=line.split(',')
    keywords.append((a,int(b)))
print(keywords[:3])
```

```python
import random
import pytagcloud
import webbrowser
from collections import Counter
```

```python
counter = Counter()
for keyword in keywords:
    counter[keyword[0]] = int(keyword[1])

ranked_tags = counter.most_common(40)
taglist = pytagcloud.make_tags(ranked_tags, maxsize=120)
pytagcloud.create_tag_image(taglist, 'wordcloud.jpg', size=(900, 600), fontname='NanumGothic', rectangular=False)
```
